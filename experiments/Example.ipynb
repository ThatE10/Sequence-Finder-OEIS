```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [3.50, 3.50]
plt.rcParams["figure.autolayout"] = True

def f(x):
    r = []
    for i in x:
        if i<=2:
            r.append((i**2)/2)
        else:   
            r.append(i)
    return r
        
   

x = np.linspace(0, 10, 100)

plt.plot(x, f(x), color='red')

plt.show()
plt.savefig('ceiledf.png')
```
