```python
def determinate(m1,sign=1):
    #This if statement compares if it is the base case or not
    #Due to the recursive nature of the function it will return the determinate of the most basic matrix
    if len(m1)==1:
        return m1[0][0]
    elif len(m1)==2:
        return m1[0][0]*m1[1][1]-m1[1][0]*m1[0][1]
    
    else:
        ##############################################
        #finds the the row or col with the most amount of zeros
        
        #[largest row/col,largest amount of 0]
        col_=[0,0]
        row_=[0,0]
        
        
        #These nested for loops first go through the first row and first column then it will go through the index column and index row
        #this is to check for any zeros within the matrix would be best optimised for it.  
        for index in range(0,len(m1)):
            col_zero_count=0
            row_zero_count=0
            #Checks for the largest amount of zeros at the individual row & col at index
            for index_individual in range(0,len(m1)):
                if m1[index][index_individual] ==0:
                    col_zero_count+=1
                if m1[index_individual][index] ==0:
                    row_zero_count+=1
                    
            #compares against the largest amount and updates it 
            if (col_[1]<col_zero_count):
                col_[1]=col_zero_count
                col_[0]=index
            
            if (row_[1]<row_zero_count):
                row_[1]=row_zero_count
                row_[0]=index
        #This is line is debugging purposes
        #print(col_)
        #print(row_)
        
        
        ##########################################
        #calculate the determinate for the row with the most amount of zeros.  
        #The sign is a variable that is introduced to ensure that the sign is flipped when required
        total=0
        #this determines if the column or row has more zeros
        if col_[1]>=row_[1]:
            
            if col_[0]%2==0:
                sign = sign*-1
            for i in range(0,len(m1)):
                if i % 2 == 0:
                    value = 1*sign
                else:
                    value = -1*sign
            
                
                if m1[col_[0]][i] == 0:
                    amount = 0
                else:
                    
                    amount = (value)*m1[col_[0]][i]*determinate(isolate_determinate(m1,col_[0],i),value)
                    
                total+=amount
        
        else:
            if row_[0]%2==0:
                sign = sign*-1
            for i in range(0,len(m1)):
                if i % 2 == 0:
                    value = 1*sign
                else:
                    value = -1*sign
                
                if m1[i][row_[0]] == 0:
                    amount = 0
                else:
                    amount = (value)*m1[i][row_[0]]*determinate(isolate_determinate(m1,i,row_[0]),value)
                total+=amount
        
        return total

def isolate_determinate(m1,i,j):
    #i is the row, j is the column
    determinate_matrix = []
    for row_i in range(0,len(m1)):
        if row_i == i:
            continue
        new_row =[]
        for col_j in range(0,len(m1[0])):
            if col_j == j:
                continue
            new_row.append(m1[row_i][col_j])
        determinate_matrix.append(new_row)
    
    return determinate_matrix
    
```
